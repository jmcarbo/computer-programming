Programa para calcular el Índice de Masa Corporal

$IMC = peso(Kg)/ estatura^2 (m2)$

Emitiremos un diagnóstico:

* Si IMC < 18.5 bajo peso
* Si IMC entre 18.5 y 24.9 peso normal
* Si IMC entre 25.0 y 29.9 sobrepeso
* Si IMC >29.9 obesidad

```
from graphviz import Digraph
dot = Digraph("IMC")
dot.node("P", "Introduce peso en Kg", shape="box")
dot.node("A", "Introduce altura en m", shape="box")
dot.node("IMC", "Calcula IMC=peso/altura^2", shape="box")
dot.node("IMC?", shape="diamond")
dot.node("BP", "Bajo peso", shape="box")
dot.node("PN", "Peso normal", shape="box")
dot.node("SP", "Sobrepeso", shape="box")
dot.node("O", "Obesidad", shape="box")
dot.edge("P", "A")
dot.edge("A", "IMC")
dot.edge("IMC", "IMC?")
dot.edge("IMC?", "BP", "<=18.5")
dot.edge("IMC?", "PN", ">18.5 Y <=24.9")
dot.edge("IMC?", "SP", ">24.9 Y <=29.9")
dot.edge("IMC?", "O", ">29.9")
dot.node("Inici", shape="circle")
dot.node("Fi", shape="circle")
dot.edge("Inici","P")
dot.edge("BP", "Fi")
dot.edge("PN", "Fi")
dot.edge("SP", "Fi")
dot.edge("O", "Fi")

dot
```
