## Llenguatges de programació

Hi ha dos grans tipus de llenguatges de programació que ens permeten interactuar amb els ordinadors: llenguatges declaratius i llenguatges procedurals.

En un llenguatge declaratiu, com pot ser un llenguatge d'interrogació d'una base de dades, li indiquem a l'ordinador què volem obtenir però no com obtenir-ho.

Per contra, en un llenguatge procedural, li indiquem exactament quins passos s'han de seguir per obtenir un resultat, com si es tractés d'una recepta de cuina.


## Diagrames de fluix

Els diagrames de fluix són una manera gràfica de representar processos i, per tant, útils també per representar gràficament programes informàtics de manera procedural. En medecina s'utilitzen sovint en els protocols hospitalaris per resumir el protocol d'actuació escrit.


```
#@markdown ### Elements bàsics del diagrama de flux
#@markdown #### Nodes
from graphviz import Digraph
dot = Digraph("Elements")
dot.node("Rectangle", shape="box")
dot.node("Rombus", shape="diamond")
dot.node("Cercle", shape="circle")  
dot
```

El rectangle serveix per representar accions, el rombus preguntes i el cercle indica l'inici i el fi del procediment.

```
#@markdown #### Enllaços
from graphviz import Digraph
dot = Digraph("Elements")
dot.node("MF", "Mirar per la finestra", shape="box")
dot.node("Tancar", "Tancar la finestra", shape="box")
dot.node("Obrir", "Obrir la finestra", shape="box")
dot.node("Fa sol?", shape="diamond")
dot.node("Inici", shape="circle")
dot.node("Fi", shape="circle")
dot.edge("Inici", "MF") 
dot.edge("MF", "Fa sol?") 
dot.edge("Fa sol?", "Tancar", "no")
dot.edge("Fa sol?", "Obrir", "sí")
dot.edge("Tancar", "Fi")
dot.edge("Obrir", "Fi")
dot
```
